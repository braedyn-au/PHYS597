```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
monSkin = np.asarray([95.311,96.326,95.404,96.928,])
monSkinSTD = np.array([0.2,0.1,0.2,0.8])
```

```python
tuesSkin = np.array([95.9,95.1,98.2,97.9])
tuesSkinSTD = np.array([0.3,0.1,0.2,0.3])
```

```python
wedSkin = np.array([98,96.4,98.7,96.8])
wedSkinSTD = np.array([1,0.2,0.2,0.1])
```

```python
thursSkin = np.array([92.5,95.2,97.1,96.3])
thursSkinSTD = np.array([0.1,0.1,0.3,0.4])
```

```python
friSkin = np.array([94.3,97.5,96.1,97.5])
friSkinSTD = np.array([0.2,0.1,0.1,0.2])
```

```python
time = [9,13,20,24]
```

```python
plt.errorbar(time,monSkin, xerr = 1, yerr=monSkinSTD, fmt='o-', ecolor='black')
plt.grid(True)
plt.xlabel("Time (hours)")
plt.ylabel("Temperature (F)")
plt.title("Monday")
plt.savefig("./mondaySkin.png", dpi=200)
```

```python
plt.errorbar(time,tuesSkin, xerr = 1, yerr=tuesSkinSTD, fmt='o-', ecolor='black')
plt.grid(True)
plt.xlabel("Time (hours)")
plt.ylabel("Temperature (F)")
plt.title("Tuesday")
plt.savefig("./tuesSkin.png", dpi=200)
```

```python
plt.errorbar(time,wedSkin, xerr = 1, yerr=wedSkinSTD, fmt='o-', ecolor='black')
plt.grid(True)
plt.xlabel("Time (hours)")
plt.ylabel("Temperature (F)")
plt.title("Wednesday")
plt.savefig("./wedSkin.png", dpi=200)
```

```python
plt.errorbar(time,thursSkin, xerr = 1, yerr=thursSkinSTD, fmt='o-', ecolor='black')
plt.grid(True)
plt.xlabel("Time (hours)")
plt.ylabel("Temperature (F)")
plt.title("Thursday")
plt.savefig("./thursSkin.png", dpi=200)
```

```python
plt.errorbar(time,friSkin, xerr = 1, yerr=friSkinSTD, fmt='o-', ecolor='black')
plt.grid(True)
plt.xlabel("Time (hours)")
plt.ylabel("Temperature (F)")
plt.title("Friday")
plt.savefig("./friSkin.png", dpi=200)
```

```python
avgSkin = np.sum([monSkin,tuesSkin,wedSkin,thursSkin,friSkin], axis=0)/5
```

```python
avgSkin
```

```python
avgSkinSTD = np.sum([monSkinSTD,tuesSkinSTD,wedSkinSTD,thursSkinSTD,friSkinSTD], axis=0)/5
```

```python
plt.errorbar(time,avgSkin, xerr = 1, yerr=avgSkinSTD, fmt='o-', ecolor='black')
plt.grid(True)
plt.xlabel("Time (hours)")
plt.ylabel("Temperature (F)")
plt.title("Week-long Average")
plt.savefig("./avgSkin.png", dpi=200)
```

```python
em = np.array([1,0.98,0.7,0.5,0.3,0.13])
iced = np.array([1.5,1.5,3.8,6.5,13,23])
icedE = np.array([0.5,0.5,1,1,2,2])
faced = np.array([0,0.5,2,5.2,8.5,14.5])
facedE = np.array([0.2,0.5,0.5,2,2,2])
```

```python
plt.errorbar(em, iced, yerr=icedE, ecolor='black')
plt.xlabel("Emissivity")
plt.ylabel("Distance (cm)")
plt.title("Distance thermometer is held from ice to measure 0$^{\circ}$C")
plt.grid(True)
plt.savefig("iceemissivity.png", dpi=200)
```

```python
plt.errorbar(em, faced, yerr=facedE, ecolor='black')
plt.xlabel("Emissivity")
plt.ylabel("Distance (cm)")
plt.title("Distance thermometer is held from head to measure 37$^{\circ}$C")
plt.grid(True)
plt.savefig("faceemissivity.png", dpi=200)
```
